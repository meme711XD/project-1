```python
import re
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def cleaning(Reviews):
    Reviews = Reviews.lower()
    Reviews = re.sub(r"http\S+", "", Reviews)
    Reviews = re.sub(r"@\w+", "", Reviews)
    return Reviews.strip()
    
df['cleaning_text'] = df['full_text'].apply(cleaning)
df
```
